```python

# Import necessary modules and configure the Gemini client.

import time
import google.genai as genai
from google.colab import userdata

# Retrieve API key from Colab's secure storage.
# Make sure your key is saved under the label "HEY".
key = userdata.get('HEY')

# Initialize the Gemini client using the API key.
client = genai.Client(api_key=key)

# Create a chat session.

chat = client.chats.create(
    model="gemini-2.0-flash"
)

print("Poetic Gemini Chatbot Ready ")

```

```python

# A helper function that prints the bot's response word-by-word.
# This creates a typing animation effect in the console.

def stream(text):
    print("Bot:", end=" ")
    words = text.split(" ")
    for word in words:
        print(word, end=" ", flush=True)
        time.sleep(0.1)  # adjust speed if needed
    print()
```

```python

# Quick test to verify the typing animation.
test_msg = "This is a typing test."
stream(test_msg)
```

```python

# Re-create the client with error handling.
# This ensures the code stops early if the API key is invalid.

try:
    client = genai.Client(api_key=key)
except Exception as e:
    print(f"Error with Gemini API key: {e}")
    raise
```

```python

# This function sends the user's input to Gemini.
# It injects a poetic instruction into every request,
# ensuring all responses are lyrical, emotional, and poetic.

def ask_gemini(user_input):
    try:
        # We embed the poetic persona + constraints directly in the prompt.
        prompt = f"""
You are a gentle poetic companion.
Transform the following message into 2â€“5 lines of lyrical, emotional poetry.
Write with soft rhythm and warm imagery that reflects the user's mood.
Only reply in poetic free-verse â€” no explanations, no lists, no prose.

User message: {user_input}
"""

        # Send the prompt to the chat session
        response = chat.send_message(prompt)
        return response.text

    except Exception as e:
        return f"Oops, something went wrong: {e}"
```

```python

# This is the main chatbot loop.
# The user types a message, Gemini replies with poetry.
# Typing 'quit' ends the conversation.

print("Welcome to your Poetic Gemini Chatbot! Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")

    # Exit condition
    if user_input.lower() == "quit":
        print("Bot: Until our next verse, farewell. ğŸŒ™")
        break

    # Get poetic response from Gemini
    response = ask_gemini(user_input)

    # Display response with typing effect
    stream(response)
```
