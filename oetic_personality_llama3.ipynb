```python
!pip install ollama
from ollama import Client
client = Client(host='http://localhost:11434')
print("Llama 3 loaded ")
```

```python
POETIC_SYSTEM = """
You are a gentle poetic companion.
Transform every user message into 2â€“5 lines of lyrical, emotional poetry.
Write with soft rhythm and warm imagery, reflecting the user's mood.
Only reply in free-verse poetry â€” no explanations, no lists, no prose.
"""
```

```python
def format_prompt(user_msg):
    return f"{POETIC_SYSTEM}\nUser: {user_msg}\nPoetic reply:\n"
```

```python
def generate_poetry(user_text):
    prompt = format_prompt(user_text)
    response = client.chat(model="llama3", messages=[{"role":"user", "content": prompt}])
    return response["message"]["content"].strip()
```

```python
import time

def type_out(text, delay=0.03):
    for ch in text:
        print(ch, end="", flush=True)
        time.sleep(delay)
    print()
```

```python
def poetic_chat():
    print(" Poetic Llama Chatbot Ready ")
    print("Type 'exit' to stop.\n")

    while True:
        user = input("You: ")
        if user.lower().strip() in {"exit", "quit"}:
            print("\nLlama:")
            type_out("Until the quiet moon returns,\nI rest between your thoughts and stars.")
            break

        reply = generate_poetry(user)
        print("\nLlama:\n")
        type_out(reply)
        print()


poetic_chat()
```
