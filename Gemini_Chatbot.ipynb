```python
import time
import google.genai as genai
from google.colab import userdata

# retrieve API key
key = userdata.get('HEY')

client = genai.Client(api_key=key)
```

```python
def stream(text):
    print('Bot:', end=' ')
    words = text.split(' ')
    for word in words:
        print(word, end=' ', flush=True)
        time.sleep(0.1)
    print()
```

```python
text = 'WE ARE BUILDING CHATBOT'
stream(text)
```

```python
try:
    client = genai.Client(api_key=key)
except Exception as e:
    print(f"Error with Gemini API key: {e}")
    raise  # stops execution if API key invalid

# create chat session
chat = client.chats.create(model="gemini-2.0-flash")
```

```python
def ask_gemini(user_input):
    try:
        prompt = (
            f"You are a comedian chatbot. Respond to every question with a funny "
            f"joke related to the topic, even if it's a serious question. "
            f"Keep it short and punchy! Here's the question: {user_input}"
        )
        response = chat.send_message(prompt)
        return response.text

    except Exception as e:
        return f"Oops, something went wrong: {e}"



```

```python
print("Welcome to your Gemini Chatbot! Type 'quit' to exit.")

while True:
    user_input = input("You: ")

    if user_input.lower() == 'quit':
        print("Bot: Bye for now!")
        break

    response = ask_gemini(user_input)
    stream(response)
```
